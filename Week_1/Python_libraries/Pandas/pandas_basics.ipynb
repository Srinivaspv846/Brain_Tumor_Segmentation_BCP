```python
# Assignment 1
'''
Import videogamesales.csv and perform the following actions on it
Add a column of 'global_sales' showing the total sales of all the different regions to the data frame and sort (highest first) and print the DataFrame according to it.
Display a plot of the total number of copies sold of each genre globally.
Filter out only the games containing 'Grand Theft Auto' in their name and display the following as a DataFrame:
a. their name
b. the platform they were released on
c. the year they were released in
d. the sum of sales in only Europe and Japan
Display a pie chart of the total sales of all Grand Theft Auto games combined in North America, Europe, Japan, and other sales.
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#loading data
df = pd.read_csv("videogamesales.csv")

#Adding 'global_sales' column
#We sum the sales from North America, Europe, Japan, and Other regions
df['global_sales'] = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum(axis=1)

#Sort by global sales (descending) and print
df_sorted = df.sort_values(by='global_sales', ascending=False)
print("Top 5 Games by Global Sales:")
print(df_sorted.head())

#Plot: Total copies sold by Genre globally
# We group by 'Genre', sum the 'global_sales', and plot
plt.figure(figsize=(12, 6))
genre_sales = df.groupby('Genre')['global_sales'].sum().sort_values(ascending=False)
genre_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Global Sales by Genre')
plt.ylabel('Copies Sold (Millions)')
plt.xticks(rotation=45)
plt.show()

#filtering out for 'Grand Theft Auto' 
gta_df = df[df['Name'].str.contains("Grand Theft Auto")].copy()

#Calculate sales for Europe + Japan
gta_df['EU_JP_Sales'] = gta_df['EU_Sales'] + gta_df['JP_Sales']

print("\nGTA Games Data (Name, Platform, Year, EU+JP Sales):")
print(gta_df[['Name', 'Platform', 'Year', 'EU_JP_Sales']])

#Pie Chart: Total GTA Sales by Region
sales_sums = gta_df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

plt.figure(figsize=(8, 8))
plt.pie(sales_sums, labels=sales_sums.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightblue', 'lightgreen', 'pink'])
plt.title('Distribution of All Grand Theft Auto Sales by Region')
plt.show()
```
